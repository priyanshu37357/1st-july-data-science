### pie

```python
x =np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x =np.array([1,2,3,4])
labels =["apple","banana","cherry","dates"]
plt.pie(x,labels=labels)
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0] #used to create an explode effect for the second slice
plt.pie(x,labels=labels,explode=myexplode)
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
plt.pie(x,labels=labels,explode=myexplode,shadow=True) #used to create a pie ch
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
plt.pie(x,labels=labels,explode=myexplode,shadow=True,startangle=90)
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
c= ['red', 'green', 'blue', 'yellow']
plt.pie(x,labels=labels,explode=myexplode,shadow=True,startangle=90,colors=c,autopct='%1.1f%%')
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple', 'banana', 'cherry', 'date']
myexplode=[0,0.2,0,0]
c= ['red', 'green', 'blue', 'yellow']
plt.pie(x,labels=labels,explode=myexplode,shadow=True,startangle=90,colors=c,autopct='%1.1f%%')
plt.legend(title='my pie chart')
plt.show()
```
